```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("all_data.csv")
```

```python
country = df.Country.unique()

df.rename(columns = {"Life expectancy at birth (years)": "LifeExpect"}, inplace = True)
df.head()
```

```python
for i in country:
    info = df[df.Country == i]
    plt.plot(info.Year, info.LifeExpect, label = i)
plt.title("Variance of Life expectancy through Years")
plt.legend()
plt.show()
```

```python
for i in country:
    info = df[df.Country == i]
    plt.plot(info.Year, info.GDP, label = i)
plt.title("Variance of Life expectancy through Years")
plt.legend()
plt.show()
```
